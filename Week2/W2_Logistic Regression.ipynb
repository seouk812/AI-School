# Logistic Regression

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold, StratifiedKFold
from pylab import scatter, show, legend, xlabel, ylabel
%matplotlib inline
```

```python
## The sigmoid function adjusts the cost function hypotheses 
## to adjust the algorithm proportionally for worse estimations
def Sigmoid(z):
    Gz = 1.0 / (1.0 + np.exp(-z))
    return Gz
```

```python
def simulateData(numData1=100, numData2=100, numData3=100):
    mu1 = [-1, -1]
    mu2 = [1, 1]

    cov1 = [[1, 0], [0, 1]]
    cov2 = [[2, 0.5], [0.5, 1]]

    data1 = np.random.multivariate_normal(mu1, cov1, numData1)
    data2 = np.random.multivariate_normal(mu2, cov2, numData2)
    print(data1)
    
    return data1, data2
```

```python
class myLogisticRegressor(object):
  
    """Implementation of a Logistic Regression"""
    
    def __init__(self, eta, epochs, numDim):
      self.eta = eta
      self.epochs = epochs
      self.theta = []
      self.initTheta(numDim)
      
    def initTheta(self, numDim):
      self.theta = np.random.rand(numDim).reshape([-1,numDim])
        
        
    def fit(self, X, y, init=True, verbose=True):
      if init==True:
        self.initTheta(len(X[1,:]))
        
      costSet = []
      for i in range(self.epochs):
          self.gradientDescent(X, y)
          cost = self.Cost_Function(X, y)
            
          costSet.append(cost)
      
      if verbose==True:
        plt.plot(range(1,len(costSet)+1), np.array(costSet), marker='o')
        plt.xlabel('Epochs')
        plt.ylabel('Cost')
        plt.show()
     
    
    def gradientDescent(self, X, y):
      CFDerivative = self.Cost_Function_Derivative(X, y)
      self.theta = self.theta - CFDerivative
    
   
    def Cost_Function_Derivative(self, X, y):
      h = self.hypothesis(X)
      print("X: "+X)
      print("Y: "+y)
      print("h: "+h)
      error = (h-y) * X
      sumErrors = np.sum(error)

      J = (float(self.eta)/float(len(y))) * sumErrors
      return J
      
      
    def Cost_Function(self, X, y):
      h = self.hypothesis(X)

      error = y*np.log(h) + (1-y)*np.log(1-h)
      sumOfErrors = np.sum(error)

      J = (-1/len(y)) * sumOfErrors
      return J   
      
      
    def hypothesis(self, X):
      z = np.dot(X, self.theta.T)
      h = Sigmoid(z.reshape(len(z),1)) 
      
      return h
    
    
    def predict(self, X):    
      return np.round(self.hypothesis(X))
    
    
    def plot_decision_regions(self, X, y, numDim, resolution=0.02):
      # setup marker generator and color map
      markers = ('s', 'x', 'o', '^', 'v')
      colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
      cmap = ListedColormap(colors[:len(np.unique(y))])

      # plot the decision surface
      x1_min, x1_max = X[:,0].min() - 1, X[:,0].max()+1
      x2_min, x2_max = X[:,1].min() - 1, X[:,1].max()+1

      xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
      
      
      data = np.array([xx1.ravel(), xx2.ravel()]).T
      if numDim>2:
        data = np.hstack((data, np.ones((len(data[:,1]), 1))))
        
      Z = self.predict(data)

      Z = Z.reshape(xx1.shape)
      plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
      plt.xlim(xx1.min(), xx1.max())
      plt.ylim(xx2.min(), xx2.max())

      # plot class samples
      labelSet = ['class 1', 'class 2', 'class 3']
      for idx, c1 in enumerate(np.unique(y)):
          plt.scatter(x=X[np.where(y==c1),0], y=X[np.where(y==c1),1], alpha=0.8, c=cmap(idx), marker=markers[idx], label=labelSet[idx])
          
      plt.show()
```

```python
clsData1, clsData2 = simulateData(100, 100)

X = np.vstack((clsData1, clsData2))
y = np.vstack((np.zeros((len(clsData1),1)), np.ones((len(clsData2),1))))

numSample, numFeature = X.shape

pos = np.where(y==1)
neg = np.where(y==0)

scatter(X[pos,0],X[pos,1],marker='o', c='b')
scatter(X[neg,0],X[neg,1],marker='x', c='r')
show()
```

```python
## make cross validation splits, taking care to choose the same
## amount of classes in each split!
CV = StratifiedKFold(n_splits=2)

## make a list of accuracy values to be populated 
acc_bias = list()
acc = list()

## this loops across all splits for the data
X_with_bias = np.hstack((X, np.ones((numSample,1))))

for train, test in CV.split(X, y):
    mylogit_bias = myLogisticRegressor(eta=0.01, epochs=1000, numDim=len(X_with_bias[1,:]))
    mylogit_bias.fit(X_with_bias[train], y[train])
    mylogit_bias.plot_decision_regions(X_with_bias[test], y[test], numDim=len(X_with_bias[1,:]))
    prediction_bias = mylogit_bias.predict(X_with_bias[test])
    
    
    mylogit = myLogisticRegressor(eta=0.01, epochs=1000, numDim=len(X[1,:]))
    mylogit.fit(X[train], y[train])
    mylogit.plot_decision_regions(X[test], y[test], numDim=len(X[1,:]))
    prediction = mylogit.predict(X[test])
    
    
    answer =  y[test]

    score_bias = np.sum(prediction_bias == answer)
    total_score_bias = float(score_bias) / float(len(X_with_bias[test]))
    acc_bias.append(total_score_bias)
    
    score = np.sum(prediction == answer)
    total_score = float(score) / float(len(X[test]))   
    acc.append(total_score)
    
print('With the bias, found mean accuracy of {:.3f}'.format(np.mean(acc_bias)))
print(acc_bias)
print('Without the bias, found mean accuracy of {:.3f}'.format(np.mean(acc)))
print(acc)
```

```python
## plot epochs vs. cost graph
mylogit_total = myLogisticRegressor(eta=0.01, epochs=1000, numDim=len(X[1,:]))
mylogit_total.fit(X, y)
mylogit_total.plot_decision_regions(X, y, numDim=len(X[1,:]))
```
